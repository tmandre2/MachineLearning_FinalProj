```python
import numpy as np

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd 
import random
import keras
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
```

```python
##IMPORT DATA
np.random.seed(1)
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
data = pd.read_csv("data/protein_only.csv") 
clin_data = pd.read_csv("data/clinical_data.csv")
# Preview the first 5 lines of the loaded data 
#data.head()
#clin_data.head()
```

```python
## Define features, datamat
feature_names=list(data)
clin_feature_names = list(clin_data)
clin_feature_names = clin_feature_names[:-3]
feature_names = feature_names + clin_feature_names

clin_datamat=np.array(clin_data)
clin_datamat = clin_datamat[:,:-3]
datamat=np.array(data)

labels = np.expand_dims(datamat[:,-1],axis=1)
datamat = np.concatenate((datamat[:,:-1],clin_datamat,labels),axis=1)
random.shuffle(datamat)

labels=datamat[:,-1]
datamat=datamat[:,:-1]
datamat = np.nan_to_num(datamat)

print(datamat.shape)
```

```python
## Normalize and ID training vs testing data
def normalize_data(datamat):
    for i in range(0, datamat.shape[1]):
        if np.max(datamat[:,i]) != 0:
            datamat[:,i]=datamat[:,i]/np.max(datamat[:,i])
    return datamat

#Split Data into Training, Validation, and Testing Sets

train_perc=0.7
test_perc=0.15
val_perc=0.15

ind1=int(datamat.shape[0]*train_perc)
ind2=ind1+int(datamat.shape[0]*val_perc)

X_train=normalize_data(datamat[:ind1,:])
y_train=labels[:ind1]

X_val=normalize_data(datamat[ind1:ind2,:])
y_val=labels[ind1:ind2]

X_test=normalize_data(datamat[ind2:,:])
y_test=labels[ind2:]
```

```python
model = Sequential()
model.add(Dense(40, input_dim=70, activation='relu'))
# model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_checkpoint = ModelCheckpoint('mybestmodel.hdf5', monitor='val_acc', 
                                   verbose=1,save_best_only=True, mode='max', period=1)

model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1, shuffle=False,
          validation_split=0.0, validation_data = (X_val,y_val), callbacks=[model_checkpoint])
```

```python
n,p = X_val.shape
pred = model.predict(X_val)
err = np.sum(np.abs(pred[:,0] - y_val))/n
print('Val error 1: {}'.format(err))
print('Val accuracy 1: {}'.format(1.-err))


model2 = load_model('mybestmodel.hdf5')

n,p = X_val.shape
pred2 = model2.predict(X_val)
err2 = np.sum(np.abs(pred2[:,0] - y_val))/n
print('Val error 2: {}'.format(err2))
print('Val accuracy 2: {}'.format(1.-err2))


n,p = X_test.shape
pred3 = model.predict(X_test)
err3 = np.sum(np.abs(pred3[:,0] - y_test))/n
print('Testing error: {}'.format(err3))
print('Test accuracy: {}'.format(1.-err3))
```
